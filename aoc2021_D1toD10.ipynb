```python
def findLifeSupportRating(matInput):
    bits = mat.shape[1]

    def findO2Rating(mat):
        matCopy = mat.copy()
        rating = 0
        for i in range(bits):
            if matCopy.shape[0] == 1: # 1 number left
                break
            else:
                # Find counts of 0, 1 in each column
                col = matCopy[:, i]
                # print(col)
                counts = np.bincount(col)
                if counts[1] > counts[0]: # 1 is most common bit
                    indexes = [k for k, b in enumerate(col) if b]
                elif counts[0] > counts [1]: # 0 is most common bit
                    indexes = [k for k, b in enumerate(col) if not b]
                else: # 1 and 0 equally common
                    indexes = [k for k, b in enumerate(col) if b]

                matCopy = matCopy[indexes, :]
                # print("Numbers left: ", matCopy)

        for idx, bit in enumerate(matCopy[0]):
            if bit: rating += 2**(bits - (idx + 1))
        
        return rating

    def findCO2Rating(mat):
        matCopy = mat.copy()
        rating = 0
        for i in range(bits):
            if matCopy.shape[0] == 1: # 1 number left
                break
            else:
                # Find counts of 0, 1 in each column
                col = matCopy[:, i]
                counts = np.bincount(col)
                if counts[1] > counts[0]: # 1 is most common bit
                    indexes = [k for k, b in enumerate(col) if not b]
                elif counts[0] > counts [1]: # 0 is most common bit
                    indexes = [k for k, b in enumerate(col) if b]
                else: # 1 and 0 equally common
                    indexes = [k for k, b in enumerate(col) if not b]

                matCopy = matCopy[indexes, :]

        for idx, bit in enumerate(matCopy[0]):
            if bit: rating += 2**(bits - (idx + 1))
        
        return rating

    O2Rating = findO2Rating(matInput)
    CO2Rating = findCO2Rating(matInput)

    return O2Rating*CO2Rating
```

```python
from typing import List
import numpy as np

class BingoBoard:
    def __init__(self, values: List[List[int]]):
        self.bingo = False
        self.dim = len(values)
        self.board = np.array(values)
        self.marker = np.zeros(shape=(self.dim, self.dim), 
                               dtype = np.int8)

    def mark_grid(self, num: int):
        '''
        Marks a grid cell with matching number
        '''
        r, c = np.where(self.board == num)
        if len(r) > 0 and len(c) > 0:
            self.marker[r.item(), c.item()] = 1

    def check_victory(self):
        '''
        Checks for a row or column of marked cells.
        Marks board with BINGO on win condition.
        '''
        # Check for horizontal match
        for i in range(self.dim):
            if self.marker[i, :].sum() == self.dim:
                self.bingo = True
                return True

        # Check for vertical match
        for j in range(self.dim):
            if self.marker[:, j].sum() == self.dim:
                self.bingo = True
                return True

        return False

    def compute_score(self, lastDrawn: int):
        '''
        Computes score of winning board
        '''
        sumUnmarked = np.sum(self.board) - np.sum(self.board*self.marker)
        return sumUnmarked*lastDrawn
```

```python
data = open(
    '/content/drive/MyDrive/Colab Notebooks/AOC/2021/day4.txt',
    'r').read().split("\n\n")

numbers = [int(n) for n in data[0].split(",")]
boards = []

for board in data[1:]:
    rows = board.split("\n")
    values = [[int(n) for n in (row.split())] for row in rows]
    boards.append(BingoBoard(values))
```

```python
result = []
for num in numbers:
    for board in boards:
        if not board.bingo: # board has not won
            board.mark_grid(num)
            if board.check_victory():
                result.append(board.compute_score(num))
        
print(result[0]) # part 1
print(result[-1]) # part 2
```

```python
from collections import defaultdict

def getNumOverlaps(data):
    counts = defaultdict(int)

    for line in data:
        seg = line.split('->')
        coords = [tuple(map(int, v.split(','))) for v in seg]
        from_, to_ = coords[0], coords[1]
        # print("Current point:", from_, to_)
        if from_[0] == to_[0]: # x1 = x2
            if from_[1] < to_[1]: # y1 < y2
                y_values = list(range(from_[1], to_[1]+1))
            else: # y1 >= y2
                y_values = list(range(from_[1], to_[1]-1, -1))

            for y in y_values:
                point = (from_[0], y)
                counts[point] += 1 
                
        elif from_[1] == to_[1]: # y1 = y2
            if from_[0] < to_[0]: # x1 < x2
                x_values = list(range(from_[0], to_[0]+1))
            else: # x1 >= x2
                x_values = list(range(from_[0], to_[0]-1, -1))

            for x in x_values:
                point = (x, from_[1])
                counts[point] += 1 
        
        else: # diagonal points
            x_dir, y_dir = to_[0] - from_[0], to_[1] - from_[1]
            sign_x = int(x_dir/abs(x_dir))
            sign_y = int(y_dir/abs(y_dir))
        
            x_values = list(range(from_[0], to_[0]+sign_x, sign_x))
            y_values = list(range(from_[1], to_[1]+sign_y, sign_y))

            for (x,y) in zip(x_values, y_values):
                counts[(x,y)] += 1

    return sum(v > 1 for v in counts.values())
```

```python
from typing import List

def simu(state: List[int], n: int):
    day = 1
    while day <= n:
        newState = state.copy()
        for i in range(len(state)):
            if state[i] == 0:
                newState[i] = 6
                newState.append(8)
            else:
                state[i] -= 1
                newState[i] -= 1
        if day == 1:
            day += 1
            continue
        state = newState
        day += 1
    
    return len(state)
```

```python
def simuDP(state, n):
    counts = [0] * 9 # stores counts of fish on each ith timer, i=0,1,...,8
    l = len(counts)
    for i in range(len(state)):
        if state[i] == 1:
            counts[1] += 1
        elif state[i] == 2:
            counts[2] += 1
        elif state[i] == 3:
            counts[3] += 1
        elif state[i] == 4:
            counts[4] += 1
        elif state[i] == 5:
            counts[5] += 1

    print("Initial counts:", counts)
    for i in range(1, n+1):
        num_babies = counts[0] # number of fishes with 0 timer on day i-1
        for j in range(l-1):
            counts[j] = counts[j + 1]
        counts[8] = num_babies
        counts[6] += num_babies
        # print("Counts on", str(i) + "th day:", counts)

    return sum(counts)

simuDP(inputState, 256)
```

```python
# The best position, a, minimizes the mean absolute distance
# between all positions and a. This is simply the median position.
def findMinFuel(positions):
    positions.sort()
    mid = len(positions) // 2
    if len(positions) % 2 == 0:
        median = (positions[mid] + positions[mid-1]) // 2
    else:
        median = positions[mid]

    fuel = 0
    for pos in positions:
        fuel += abs(pos - median)

    return fuel
```

```python
findMinFuel(pos)
```

```python
def findMinFuelP2(positions):
    answers = []
    for pos in range(min(positions), max(positions)):
        fuel = 0
        for p in positions:
            fuel += (abs(pos - p)*(abs(pos - p) + 1))/2
        answers.append((pos, fuel))

    return min(answers, key = lambda x: x[1])

findMinFuelP2(pos)
```

```python
def decode(sigs):
    wmap, dmap = {}, {} # map of wires -> segments, map of digits -> signals
    five_seg_sigs = [] # corresponds to digits 2,3,5
    six_seg_sigs = [] # corresponds to digits 0,6,9
    for sig in sigs.split(' '):
        numSegs = len(sig)
        if numSegs == 2:
            dmap[1] = sig
        elif numSegs == 3:
            dmap[7] = sig
        elif numSegs == 4:
            dmap[4] = sig
        elif numSegs == 5:
            five_seg_sigs.append(sig)
        elif numSegs == 6:
            six_seg_sigs.append(sig)
        elif numSegs == 7:
            dmap[8] = sig

    assert(len(five_seg_sigs) == 3)
    assert(len(six_seg_sigs) == 3)

    horz_segs = set.intersection(*list(map(set, five_seg_sigs))) # get common segments in digits 2,3,5
    # print("Horizontal segments are in:", horz_segs)

    wmap["top"] = (set(dmap[7]) - set(dmap[1])).pop() # infer top segment from 7 and 1
    wmap["mid"] = ((set(dmap[4]) - set(dmap[1])).intersection(horz_segs)).pop() # infer middle segment from 4 and horizontal segments
    wmap["bottom"] = (horz_segs - set([wmap["top"], wmap["mid"]])).pop() 
    wmap["topL"] = (set(dmap[4]) - (set(dmap[1]).union(wmap["mid"]))).pop() # upper left segment

    dmap[3] = [x for x in five_seg_sigs if set(dmap[1]).issubset(set(x))].pop() # 3 has all segments of 1
    five_seg_sigs.remove(dmap[3])
    dmap[5] = [x for x in five_seg_sigs if set(wmap["topL"]).issubset(set(x))].pop() # 5 has upper left segment
    five_seg_sigs.remove(dmap[5])
    dmap[2] = five_seg_sigs.pop()

    for sig in six_seg_sigs:
        if not set(wmap["mid"]).issubset(set(sig)): # 0 has no middle segment
            dmap[0] = sig
        elif set(dmap[4]).issubset(set(sig)): # 9 has all segments of 4
            dmap[9] = sig
        elif set(dmap[5]).issubset(set(sig)): # 6 has all segments of 5
            dmap[6] = sig

    return wmap, dmap


def decodeSum(data):
    sum = 0
    for line in data.readlines():
        signals, output = line.rstrip().split('|')
        wireMap, digitMap = decode(signals)
        
        # Invert dictionary mapping of digit -> signals
        # digitMap = {v:k for k,v in digitMap.items()}
        # print("Completed mapping of signals -> digit")
        # for k,v in digitMap.items():
        #     print(k, v)
        
        num = ''
        for sig in output.split(' '):
            if sig: 
                for k, v in digitMap.items():
                    if set(v) == set(sig): num += str(k); break
        # print("Output from signal is:", num)
        sum += int(num)
    
    return sum
```

```python
import numpy as np
data = open(
    '/content/drive/MyDrive/Colab Notebooks/AOC/2021/day9.txt',
    'r')

mat = []

for line in data.readlines():
    arr = list(line.rstrip())
    mat.append(list(map(int, arr)))

mat = np.array(mat, dtype = np.int8)
```

```python
def isLowPoint(r, c, rows, cols):
    if mat[r, c] == 9: # max height
        return False
    if r == 0:
        bottom = mat[r+1, c] # bottom adjacent point
        if c == 0: # top left
            if mat[r, c] < bottom and mat[r, c] < mat[r, c+1]: return True
        elif c == cols - 1: # top right
            if mat[r, c] < bottom and mat[r, c] < mat[r, c-1]: return True
        else:
            if mat[r, c] < bottom and mat[r, c] < mat[r, c-1] and mat[r,c] < mat[r,c+1]:
                return True
    elif r == rows - 1:
        top = mat[r-1, c] # top adjacent point
        if c == 0: # bottom left
            if mat[r,c] < top and mat[r, c] < mat[r, c+1]: return True
        elif c == cols - 1: # bottom right
            if mat[r, c] < top and mat[r, c] < mat[r, c-1]: return True
        else:
            if mat[r, c] < top and mat[r, c] < mat[r, c-1] and mat[r,c] < mat[r,c+1]:
                return True  
    else:
        top, bottom = mat[r-1, c], mat[r+1, c]
        if c == 0: # left edge point
            if mat[r,c] < top and mat[r,c] < bottom and mat[r,c] < mat[r, c+1]:
                return True
        elif c == cols - 1: # right edge point
            if mat[r,c] < top and mat[r,c] < bottom and mat[r,c] < mat[r, c-1]:
                return True
        else:
            left, right = mat[r, c-1], mat[r,c+1]
            if mat[r,c] < top and mat[r,c] < bottom and mat[r,c] < left and mat[r,c] < right:
                return True
```

```python
def findRiskLow(mat):
    rows, cols = mat.shape[0], mat.shape[1]
    risk = 0
    for i in range(rows):
        for j in range(cols):
            if isLowPoint(i, j, rows, cols): risk += (1 + mat[i,j])
    return risk
```

```python
from functools import reduce
# Idea: Starting from each low point, call dfs on adjacent points (as long as 
# it is larger) until we hit the highest point(9), keeping track of points 
# in our path that form the basin
def findTop3Basins(mat):
    rows, cols = mat.shape[0], mat.shape[1]
    def findLowPoints(mat):
        ans = []
        for i in range(rows):
            for j in range(cols):
                if isLowPoint(i, j, rows, cols): ans.append((i, j))
        return ans

    def DFS(i, j, visited):
        for nei in [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]: # top, bottom, left, right
            if nei[0] < 0 or nei[0] >= rows or nei[1] < 0 or nei[1] >= cols:
                continue
            elif nei not in visited and mat[nei] > mat[i, j] and mat[nei] < 9:
                visited.add(nei)
                DFS(nei[0], nei[1], visited)

    points = findLowPoints(mat)
    # print(points)
    basin_sizes = []
    for lp in points:
        # print("Low point:", lp)
        visited = set([lp])
        DFS(lp[0], lp[1], visited)
        # print(visited)
        basin_sizes.append(len(visited))

    return reduce(lambda x, y: x*y, sorted(basin_sizes, reverse=True)[:3])
        
```

```python
def findCorruptedLinesErrorScore(input):
    '''Idea: We need to close inner brackets before outer brackets. 
    Whenever there's a closed bracket, check and see if it pairs with 
    the last open bracket (can be tracked using a stack), if so, remove
    the bracket and process the next open bracket.
    '''
    pointsMap = {")": 3,
                "]": 57,
                "}": 1197,
                ">": 25137}
    bracket_pairs = {"(": ")", 
                    "[": "]", 
                    "{": "}", 
                    "<": ">"}
    total_score = 0
    for line in input.readlines():
        open = [] # stack to track open brackets
        for bracket in line.rstrip("\n"):
            if bracket in bracket_pairs: open.append(bracket)
            else: # closed bracket
                last_open = open[-1]
                if bracket == bracket_pairs[last_open]:
                    open.pop()
                else:
                    print("Corrupted line, expected", bracket_pairs[last_open] + ", but found", bracket + " instead.")
                    total_score += pointsMap[bracket]
                    break

    return total_score
```

```python
def findMiddleCompletionScore(input):
    pointsMap = {")": 1,
                "]": 2,
                "}": 3,
                ">": 4}
    bracket_pairs = {"(": ")", 
                    "[": "]", 
                    "{": "}", 
                    "<": ">"}
    completion_scores = []
    for line in input.readlines():
        open = [] # stack to track open brackets
        corrupted = False
        for bracket in line.rstrip("\n"):
            if bracket in bracket_pairs: open.append(bracket)
            else: # closed bracket
                last_open = open[-1]
                if bracket == bracket_pairs[last_open]:
                    open.pop()
                else:
                    corrupted = True
                    # print("Corrupted line, expected", bracket_pairs[last_open] + ", but found", bracket + " instead.")
                    break

        if not corrupted:
            completion_string = ''
            while open:
                completion_string += bracket_pairs[open.pop()]

            print(line.rstrip("\n") + " is completed by adding", completion_string)
            tot_score = 0
            for s in completion_string:
                tot_score = tot_score*5 + pointsMap[s]

            print("Completion string has score", tot_score)
            completion_scores.append(tot_score)
    
    mid = len(completion_scores) // 2
    middle_score = sorted(completion_scores)[mid]

    return middle_score
```
